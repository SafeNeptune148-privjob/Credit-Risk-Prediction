# Let's deal with null values

```python
df.isnull().sum()
```

#### Let's see if the null values that some features have comes from the same row

```python
missing_one = df.isnull().sum()
missing_one = missing_one[missing_one == 1]
missing_one
```

```python
common_missing = set(df[df[missing_one.index[0]].isnull()].index)
for col in missing_one.index[1:]:
    common_missing &= set(df[df[col].isnull()].index)
print("The row with the missing values are: ", common_missing)
```

```python
#We are going to drop the row that is causing trouble with only ne missing value
df.drop(list(common_missing)[0], inplace = True)
```

```python
df.isnull().sum()
```

```python
#Let's go with the annuity
print(df["AMT_ANNUITY"].dtype)
# We will impute it with the median
```

```python
sns.histplot(df["AMT_ANNUITY"].dropna(), kde=True)
plt.title("Distribution of Annuity")
plt.show()
```

```python
#We fill the missing values with the median
df["AMT_ANNUITY"] = df["AMT_ANNUITY"].fillna(df["AMT_ANNUITY"].median())
df.isnull().sum()
```

```python
#Let's go with the occupation type
plt.figure(figsize = ( len(df["OCCUPATION_TYPE"].unique()),15))
sns.histplot(data = df, x = df["OCCUPATION_TYPE"], hue = df["TARGET"])
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

```python
#Let's see the proportion of missing values
print(df["OCCUPATION_TYPE"].isnull().sum()/len(df))
```

```python
np.random.seed(42)

for target in df["TARGET"].unique():
    mask = (df["TARGET"] == target) & (df["OCCUPATION_TYPE"].isna())
    n_missing = mask.sum()
    probs = df.loc[
        (df["TARGET"] == target) & (df["OCCUPATION_TYPE"].notna()),"OCCUPATION_TYPE"
    ].value_counts(normalize = True)
    imputed_values = np.random.choice(probs.index, size = n_missing, p = probs.values)
    df.loc[mask, "OCCUPATION_TYPE"] = imputed_values
```
